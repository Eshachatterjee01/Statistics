```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

A KDE plot is a smooth, non-parametric estimate of the probability density function based on observed data, not the true PDF.

```python
sns.kdeplot(data=df['sepal_length']) #dekh ke normal distribution lag raha hai..par pata kaise chalega see down
```

```python
#yahape sepal length ke sare values ko nikalke sort kar diya.
temp = sorted(df['sepal_length'].tolist())
```

```python
temp
```

```python
len(temp)
```

```python
#1 se leke 100 tkka loop chalake sare ke sare percentile calculate kiye.1 percentile 2nd percentile .... 100 percentile.
#You are calculating all percentiles (1% to 100%) of the data temp and storing them in a list called y_quant.
y_quant = []
for i in range(1,101):
  y_quant.append(np.percentile(temp,i))
```

Generating a normally distributed data as a sample

```python
sample = np.random.normal(loc=0,scale=1,size=1000)
```

```python
#again we will calculate percentile for the sample data also

x_quant = []
for i in range(1,101):
  x_quant.append(np.percentile(sample,i))
```

```python
#dono ke bich me scatter plot plot karna hai

sns.scatterplot(x=x_quant,y=y_quant)
```

Using model also you can check --> using stats model.ye sab khudse kar lega.bas ye batani hai ki line kaisi draw karni hai.45 degree ka ya konsa standardized line,regression line you can check from sm.qqplot from internet.

```python
# using statsmodel
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Create a QQ plot of the two sets of data
fig = sm.qqplot(df['sepal_length'], line='45', fit= True)

# Add a title and labels to the plot
plt.title("QQ Plot")
plt.xlabel('Theoretical Quantiles')
plt.ylabel("Sample Quantiles")

#show the plot
plt.show()
```

So,we can saythis is sort of normal but not full fledged normal distribution,not a perfect normal distribution.
